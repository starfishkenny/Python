```python
# 재무데이터 전처리 함수화
def get_finance_data(path):
    data_path = path
    raw_data = pd.read_excel(data_path)
    big_col = list(raw_data.columns)
    small_col = list(raw_data.iloc[0])
    
    new_big_col = []
    for num,col in enumerate(big_col):
        if 'Unnamed' in col:
            new_big_col.append(big_col[num-1])
        else:
            new_big_col.append(big_col[num])
    
    raw_data.columns = [new_big_col, small_col]
    clean_df = raw_data.loc[raw_data.index.dropna()]
    
    return clean_df
```

```python
# 재무비율 데이터와 투자지표 데이터 역시 똑같은 문제가 있으므로
# get_finance_data 함수를 활용하여 가져오면 됨

fs_path = r'E:\jupyter\Quant with Python\Excel_data\재무제표데이터.xlsx'
fs_df = get_finance_data(fs_path)

fr_path = r'E:\jupyter\Quant with Python\Excel_data\재무비율데이터.xlsx'
fr_df = get_finance_data(fr_path)

invest_path = r'E:\jupyter\Quant with Python\Excel_data\투자지표데이터.xlsx'
invest_df = get_finance_data(invest_path)
```

* 저 PER, 고 ROA 추구
* PER은 주식의 저평가 정도를 알아보는 가장 대표적인 지표
* 그러므로 상대적으로 PER이 작은 주식을 선호

```python
# 2020/09 기준으로 PER 작은 순으로 나열
invest_df.sort_values(by=('2020/09','PER'))
```

```python
# 데이터 형태를 맞춰야 제대로 정렬이 일어날 것
# 숫자열로 바꾼 후 정렬해줘야 함
invest_df[('2020/09','PER')] = pd.to_numeric(invest_df[('2020/09','PER')])
per_sorted = invest_df.sort_values(by=('2020/09','PER'))
```

```python
per_sorted['2020/09']
```

```python
# 고 ROA -> 내림차순으로 정렬
fr_df[('2020/09','ROA')] = pd.to_numeric(fr_df[('2020/09','ROA')]) # 에러 발생
```

```python
# 에러 해결 -> ROA 칼럼의 'N/A(IFRS)'를 NaN으로 바꾸기
def check_IFRS(x):
    if x=='N/A(IFRS)':
        return np.NaN
    else:
        return x

fr_df[('2020/09','ROA')] = pd.to_numeric(fr_df[('2020/09','ROA')]).apply(check_IFRS)
```

```python
# ROA 칼럼 기준으로 내림차순 정렬
fr_df[('2020/09','ROA')] = pd.to_numeric(fr_df[('2020/09','ROA')])
sorted_roa = fr_df.sort_values(by=('2020/09','ROA'), ascending=False)
```

```python
# PER 오름차순 함수 정리
def low_per(invest_df,index_date,num): # 투자지표 데이터, 기준 날짜, 반환할 주식의 개수를 인자로 받음
    invest_df[(index_date,'PER')] = pd.to_numeric(invest_df[(index_date,'PER')])
    per_sorted = invest_df.sort_values(by=(index_date,'PER'))
    return per_sorted[index_date][:num]
```

```python
# 예시
low_per(invest_df,'2020/09',10)
```

```python
# ROA 내림차순 함수 정리
def high_roa(fr_df,index_date,num):
    fr_df[(index_date,'ROA')] = fr_df[(index_date,'ROA')].apply(check_IFRS)
    fr_df[(index_date,'ROA')] = pd.to_numeric(fr_df[(index_date,'ROA')])
    sorted_roa = fr_df.sort_values(by=(index_date,'ROA'), ascending=False)
    return sorted_roa[index_date][:num]
```

```python
# 예시
high_roa(fr_df,'2020/09',10)
```

#### 저평가된 주식들
* PER 말고도 PBR, PSR, PCR과 같은 지표가 해당 주식의 저평가 정도를 알아보기 위해 많이 사용됨
* PER이 순이익 대비 주가의 저평가 정도를 나타내는 지표인 반면, PBR은 순자산 대비 주가의 저평가 정도를 나타냄
#### F-Score 
* 조셉 피오트로스키가 만든 F-Score는 우량 기업의 주식을 찾는 대표적인 방법
* 다음과 같은 조건들을 설정하고 조건이 충족하면 1점을 주고 아니면 0점을 주어서 총 점수를 계산함
  - 당기순이익이 0 이상인가?
  - 영업현금흐름이 0 이상인가?
  - ROA가 전년 대비 증가했는가?
  - 영업현금흐름이 순이익보다 높은가?
  - 부채비율이 전년 대비 감소했는가?
  - 유동비율이 저년 대비 증가했는가?
  - 당해 신규주식 발행을 하지 않았는가?
  - 매출총이익이 전년 대비 증가했는가?
  - 자산회전율이 전년 대비 증가했는가?
#### 모멘텀
* 추세추종 전략이라고도 함

* 모멘텀 전략을 구현하는 가장 쉬운 방법 $\rightarrow$ 최근 n개월의 수익률을 계산하여 이를 종목들끼리 서로 비교하는 방법일 것
