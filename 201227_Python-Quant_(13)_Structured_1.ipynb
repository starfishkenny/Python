### 펀드와 구조화 상품 개념과 설계

#### 구조화 상품
+ 주식, 채권, 선물, 옵션 등을 기본 상품이라고 한다면,
+ 구조화 상품은 기본 상품들을 조합하여 새로운 수익 구조를 만들어낸 것
+ ETF, ETN, ELW, ELS, DLS, ELD 등

#### 구조화 상품 용어
+ 액티브 vs. 패시브
  - 액티브: 벤치마크 지수 이상의 수익률을 목표로 삼고 그때그때 펀드매니저의 판단에 따라 다양한 운용을 방법을 선택
  - 패시브: 벤치마크 지수를 기준으로 운용 방식을 사전에 정의하고 정해진 수익률에 따라가도록 설계해 펀드매니저의 개입 없이 자동으로 운용
+ 알파 vs. 베타
  - 알파: 금융상품을 운용할 때 비교 대상보다 높은 수익을 추구하는 것
  - 베타: 비교 대상의 수익률과 유사한 패턴의 수익을 추구하는 것
  - 주로 액티브 펀드는 알파, 패시브 펀드는 베타를 추구
+ 벤치마크
  - 패시브 펀드 중 특정 지수(인덱스)를 따라가도록 설계된 펀드를 인덱스 펀드라고 함
  - "KOSPI200 대비 x% 초과수익 또는 KOSPI200 만큼의 수익을 내겠다"는 광고에서...
  - KOSPI200이 벤치마크(기준점)가 됨
  - 주로 주가지수를 선정하는데 여러 지수를 섞어 새로운 지수(customized benchmark index)를 만들어 벤치마크로 삼기도 함

#### ETF (Exchange Traded Fund)
+ 낮은 수수료
  - ETF는 대부분 인덱스 펀드로 액티브 펀드보다 운용하는 데 수고가 덜 할 수 있음
+ 실시간 매매
  - 주식처럼 거래소에서 실시간 거래 가능
  - 거래소는 ETF의 유동성을 보자아하기 위해 LP(유동성공급자)라는 장치 有
    * 인기가 없는 종목들은 증권사를 지정하여 강제로 매도 매수 주문을 내도록 하는 것
+ NAV
  - AUM(Asset Under Management): ETF가 담고 있는 총자산(주식, 채권, 현금 등)
  - AUM / 총 펀드 좌수 = NAV(Net Asset Value: 순자산가치) (펀드에서는 '주' 대신 '좌'라는 표현 사용)
  - 거래가 진행되고 있는 장 중에도 NAV 추정치를 발표 -> iNAV(Inferred Net Asset Value)
+ 괴리율
  - 주식과 마찬가지로 과열되거나 저평가될 수 있음
  - 거래소는 투자자들이 ETF를 적정한 가격에 매매할 수 있도록 정상 가격을 알려줌
  - => ETF의 정상가 = NAV, 괴리율 = 시장에서 매매되는 ETF 가격과 NAV의 차이의 비율
  - 운용사가 관리를 잘하는 ETF일수록 괴리율이 작음

#### 펀드 설계 및 운용
1. 벤치마크 선택
2. 펀드 스타일 결정
  + 지수형, 레버리지형(X2, X3), 인버스형(-1배, -2배), 초과수익형(커버드콜)
3. 펀드 구조화
  + 원하는 수익률을 달성하기 위해 필요한 운용 방식을 결정
4. 편입종목 및 비중 결정
5. 투자자 모집
6. 펀드 설정
7. 리밸런싱
  + 투자자금이 늘고 줄어듦에 대응하여 편입한 주식 매수/매도
8. 펀드 환매


- 자산운용사가 실제로 펀드를 운용할 때는 비용이 들어가기 때문에 자산을 모두 주식매입에 사용할 수는 없음
  + 자산의 99% 정도만 주식 매입에 사용하고 1%는 투자자가 환매를 신청하는 등 이벤트에 대응하기 위해 현금으로 보유
  + 이러한 차이는 펀드 수익률과 지수 수익률에 오차가 발생하는 원인이 됨
    * 이러한 오차를 '추적오차율'이라고 함
