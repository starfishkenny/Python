```python
# 그래프 모양 설정하기
# color:색상 / linestyle: 선 모양 / marker: 표시 모양
# 'b': blue / 'g': green / 'r': red / 'c': cyan / 'm': magenta / 'y': yellow / 'k': black / 'w': white
# '--' / '-' / ':' / '-.'
# '.': 점 / 'o': 원 / '8': 팔각형 / 's': 사각형 / 'p': 육각형 / '*': 별 / '+': 더하기 / 'x': 곱하기 / 'D': 다이아몬드
# 이외에도 마커사이즈 markersize='숫자' / 마커 색 markerfacecolor='색상' / 
# 마커 경계색 markeredgecolor='색상' / 투명도 alpha='0~1사이 소수' 등도 있음
# ref: https://matplotlib.org/api/index.html
# 색은 빨강, 모양은 점선, 마커는 원인 그래프 그리기
plt.plot(x,y,color='r',linestyle=':',marker='o') # plt.plot(x,y,'ro--')도 같음
```

```python
# 제목, 축 이름, 범례 삽입
plt.plot(x,y,color='m',linestyle=':',marker='*',markersize='20')
plt.title('제목')
```

```python
# 여러 개의 그래프를 모아서 보여주기
fig = plt.figure(figsize=(16,10)) # 그래프 사이즈를 지정하고 fig에 할당

# subplot 지정: add_subplot(행,열,순서)
ax1 = fig.add_subplot(2,3,1) # 2행 3열 fig 중 첫째 자리를 ax1에 할당
ax1.plot(x,y) # ax1에 plot() 그리기

ax2 = fig.add_subplot(2,3,2) # 2행 3열 fig 중 둘째 자리를 ax2에 할당
ax2.hist(y) # ax2에 hist() 그리기

ax3 = fig.add_subplot(2,3,3) # 2행 3열 fig 중 셋째 자리를 ax3에 할당
ax3.scatter(x,y) # ax3에 scatter() 그리기

ax4 = fig.add_subplot(2,3,4) # 2행 3열 fig 중 넷째 자리를 ax4에 할당
ax4.plot(x,y,color='r',linestyle='--')
```

```python
# 다른 방법 -> 껍데기를 fig에 할당하는 동시에 axes에도 할당 / axes는 껍데기인 figure 안에 들어 있는 각각의 그래프 하나씩을 가리킴
# fig, axes = plt.subplots(n,m) => figure 껍데기 안에 n행 m열의 subplot을 넣을 것이고 각 subplot을 axes에 할당한다는 것

fig, axes = plt.subplots(2,3,figsize=(16,10))

axes[0,0].plot(y)
axes[0,1].hist(y)
axes[0,2].scatter(x,y)
axes[1,0].plot(y, 'g--')
```
