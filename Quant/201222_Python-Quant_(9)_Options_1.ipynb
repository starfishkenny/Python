### 옵션 거래
#### 옵션 거래의 개념
  + 옵션만으로도 내용이 방대
  + 옵션가격을 추정하는 방법만 여러가지...
    - 대표적으로 블랙-숄즈 모형, 유한차분법, 몬테카를로 시뮬레이션 등
    - 여기서는 블랙-숄즈 방정식을 활용해본다

#### 권리에 대한 계약
  + 일반적인 투자자로서 옵션을 이해하기 어려움
    - 실물을 대상으로 거래 X, 금융상품을 거래할 '권리'를 거래하는 것이기 때문
    - 옵션: 약간의 돈을 내고 주식 매매를 예약해두었다는 뜻
  + 현물/선물/옵션 거래의 비유
    - 현물거래: 편의점에서 상품을 구매 (현재시점에 물건을 사고파는 행위)
    - 선물거래: 아파트 매수를 위해 1월에 매도자와 계약하고 계약금 10% 지불 후 잔금일인 3월에 남은 90%를 치르고 등기 이전을 하는 것
    - 옵션거래: 크리스마스 파티를 위해 예약금 10만원으로 레스토랑 예약
        * 10만원: 레스토랑 사용 권리를 위한 대가
        * 예약자가 사정이 생겨 파티를 못한다면 10만 원 포기하고 예약 취소 가능
        * 레스토랑은 예약금을 받았으므로 크리스마스에 레스토랑을 에약자에게 제공할 의무를 짐
        * 매수자 -> 예약자 / 기초자산 -> 레스토랑 이용권 / 만기 -> 크리스마스 인 '옵션거래'

| 상품    | 주식(현물)    | 선물    | 옵션    |  
|:--- | ---: | :---: | :---:|  
| 가격         | 현재 가격    | 현재 가격    | 합의한 가격(행사가)   |  
| 이행 시기     | 지금    | 미래    | 미래   |
| 개념         | 매매    | 계약    | 예약   |
| 의무         | 해당 없음    | 매수자:O 매도자:X    | 매수자:X 매도자:O   |

#### 옵션의 가치
  + 시간가치: 계약일로부터 만기일까지 기간이 길수록 계약의 가치를 변동시킬 수 있는 사건이 발생할 가능성 높음
    - 시간가치는 시간이 지날수록 줄고, 만기가 되면 사라짐
  + 내재가치: 지금 현재가 만기라고 가정할 때 옵션의 가치가 얼마나 되는지를 측정
    - 가능성의 가치 배제
    - ATM (at the money: 등가격) -> 내재가치 0인 상태
    - ITM (in the money: 내가격) -> 내재가치 + ==> 기초자산의 가격이 권리 행사 구간 내에 有
    - OTM (out of the money: 외가격) -> 내재가치 - ==> 기초자산의 가격이 권리 행사 구간 밖에 有
  + 옵션의 가치 = 내재가치 + 시간가치

#### 변동성
  + 변동성(volatility): 시장 상황의 변동으로 인한 가능성의 크기
    - 변동성이 커지면, 즉 가능성이 커지면 시간가치가 커져서 옵션가격이 올라감

#### 콜 vs. 풋
  + 콜: 만기 시 살 수 있는 권리
  + 풋: 만기 시 팔 수 있는 권리
  + 현 시점의 포지션은 매도/매수, 만기 시점의 포지션은 콜/풋으로 구분하여...
    - 콜 매수 (long call) / 콜 매도 (short call) / 풋 매수 (long put) / 풋 매도 (short put)
  + 매수 포지션: 프리미엄을 지불하고 옵션을 구매한 측
    - 유리한 경우 권리 행사 / 불리한 경우 권리 포기
    - 콜 매수는 싸게 살 권리 => 현물가격 > 행사가 ==> 수익 발생
    - 풋 매수는 비싸게 팔 권리 => 현물가격 < 행사가 ==> 수익 밣생
  + 매도 포지션: 프리미엄을 받고 옵션을 판매한 측
    - 매수 포지션과는 달리 권리는 없고 의무에 대한 책임만 있으므로 위험성 大
    - 콜 매도는 싸게 팔 의무 => 현물가격 > 행사가 ==> 손해 발생
    - 풋 매도는 비싸게 살 의무 => 현물가격 < 행사가 ==> 손해 발생
    - 의무만 있는 매도 포지션을 잡는 이유...
      * 프리미엄 수익! -> 안정적인 수익 / 복권 발행자를 생각하면 이해하기 쉬움 (복권 수익 vs. 당첨금)

#### 다양한 옵션 상품
  1. 다양한 가격
  2. 매달 만기
  3. ELW (Equity Linked Warrant): 미니옵션보다 승수를 낮게 책정한 옵션
    - 소매용 상품이라 소비자를 위한 안전장치 추가 => 매수 포지션만 가능 (콜 매수, 풋 매수)
    - 매도는 기관인 LP(Liquidity Provider)가 전담
  4. 미국형 옵션, 이색 옵션 (조기 종료형 옵션 등...)

#### 옵션 이론가: 블랙-숄즈 방정식
$ C(콜) = S_0N(d_1)-Ke^{-rT}N(d_2) $ \
$ P(풋) = Ke^{-rT}N(-d_2)-S_0N(-d_1) $ \
\
$ d_1 = {{ln{S_0 \over K}} + (r+{\sigma^2 \over 2})T \over \sigma \sqrt T} $ \
$ d_2 = d_1 - \sigma \sqrt T = {{ln{S_0 \over K}} + (r - {\sigma^2 \over 2})T \over \sigma \sqrt T} $

+ $S_0$: 기초자산 가격 (stock price at time 0)
+ K: 행사가격 (strike price)
+ T: 잔존만기<sub>(연율로 환산)</sub> (time to maturity)
+ r: 이자율<sub>(연속복리)</sub> (continuosly compounded risk-free rate)
+ $\sigma$: 기초자산 변동성 (volatility)
+ $N(x)$: 표준정규분포의 누적분포함수
\\
+ 풋옵션은 콜옵션과 반대의 포지션이므로 만기에 '행사가격 - 기초자산 가격) 만큼의 수익을 실현
  - 그러므로 콜옵션의 가격과 반대로 계산

+ $N(d_1)$, $N(d_2)$ -> 옵션이 행사될 기댓값
+  블랙-숄즈 모형에서...
  - $S_N(d_1)$ = 기초자산의 기댓값
  - $Ke^{-rT}N(d_2)$ = 행사가격의 기댓값 ($Ke^{-rT}$는 행사가격의 현재가치)

#### 옵션 그릭스(Greeks)
  + $\Delta$(델타): 기초자산 가격 변동 대비 옵션 가격 변동 (블랙-숄즈 방정식을 S로 편미분)
    - 델타가 0.5라면 기초자산 가격이 1% 변동할 때, 옵션가격은 0.5% 변동한다는 것을 의미
    - $\Delta$ = $ dC \over dS $ = $\Delta(call)$ = $N(d_1)$ , $\Delta(put)$ = $N(d_1)-1$
    - 콜옵션 매수 포지션의 경우..
      * 내가격일수록 1에 가깝고 외가격일수록 0에 가까움\
        - 행사 가능성이 높을수록 기초자산 가격 변동과 비슷한 만큼 옵션가격이 움직이고
        - 행사 가능성이 낮을수록 기초자산 가격 변동에 비해 옵션가격 변동은 둔감함
  + $\Gamma$(감마): 기초자산 가격 변동 대비 델타의 변동 (델타를 S로 편미분)
    - $\Gamma$ = $ d^2C \over dS^2 $ = $ N'(d_1) \over {S_0} {\sigma} {\sqrt T} $
    - $N'(x)$ = $ {1 \over 2 \pi} e^{{-x^2} / 2} $
    - 등가격 근처에서 가장 큰 값을 가짐 = 행사가격 근처에서는 기초자산 가격의 변화에 따라 옵션가격 변화 속도가 더 빠름
  + $\Theta$(세타): 시간 경과 대비 옵션가격의 변동율 (블랙-숄즈 방정식을 T로 편미분)
    - $\Theta(call)$ = $ - {{{S_0}N'(d_1) \sigma} \over {2 \sqrt T}} - {{rK{e^{-rT}}}N(d_2)} $
    - $\Theta(put)$ = $ - {{{S_0}N'(d_1) \sigma} \over {2 \sqrt T}} + {{rK{e^{-rT}}}N(-d_2)} $
    - T가 0에 가까울수록, 즉 만기가 가까워질수록 세타는 가파르게 떨어짐
  + $v$(베가): 기초자산 변동성($\sigma$) 대비 옵션가격의 변동율 (블랙-숄즈 방정식을 $\sigma$로 편미분)
    - $v$ = $ dC \over d \sigma $ = $ {S_0}{\sqrt T}N'(d_1) $
    - $N'(x)$ = $ {1 \over 2 \pi} e^{{-x^2} / 2} $
    - 변동성이 낮을수록 옵션가격의 민감도가 높음
  + $\rho$(로): 이자율($r$) 변동 대비 옵션가격의 변동율 (블랙-숄즈 방정식을 $r$로 편미분)
    - $\rho$ = $ dC \over dr $ = $ \rho(call) = KTe^{-rT}N(d_2) $ , $ \rho(put) = -KTe^{-rT}N(-d_2) $
    - 이자율 변동은 자주 일어나는 일이 아니므로 실제 트레이딩에서는 거의 사용되지 않는 지표
