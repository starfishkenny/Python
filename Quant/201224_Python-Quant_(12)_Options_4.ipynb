#### 잔존만기 계산

```python
def time_to_maturity(t0,T,y=252):
    return (np.busday_count(t0,T)/y)
# 잔존만기를 계산하기 위해서 영업일을 구하는 함수로 인자가 deprecated 되버린듯
# 참고: https://optimizing.tistory.com/13
```

```python
time_to_maturity('2018-08-01','2018-12-14')
```

```python
#### 블랙-숄즈 방정식을 이용한 콜옵션가격 구하기
def call_price(S, K, ttm, r, sigma):
    d1 = (np.log(S/K) + (r + sigma**2*0.5) * ttm) / (sigma*np.sqrt(ttm))
    d2 = (np.log(S/K) + (r - sigma**2*0.5) * ttm) / (sigma*np.sqrt(ttm))
    # d2 = d1 - sigma * sqrt(ttm)도 같음
    val = (S * stats.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * ttm) * stats.norm.cdf(d2, 0.0, 1.0))
    return val
```

```python
ttm = time_to_maturity('2018-09-01','2018-12-14')
r = 0.025
sigma = 0.15
```

```python
# call_price 호출 => call_price(기초자산, 행사가, 잔존만기, 이자율, 변동성)
call_price(350, 350, ttm, r, sigma)
```

#### 기초자산 변화에 따른 콜옵션가격 변화

```python
call_space = dict() # 결과값을 저장하기 위한 딕셔너리 생성
S = np.arange(200, 400, 10)
for s in S:
    call_space[s] = call_price(s, 300, ttm, r, sigma)
call_space
```

#### 기초자산도 변하고, 행사가격도 변하는 경우

```python
call_space = dict()
K = np.arange(250, 350, 10)
S = np.arange(200, 400, 10)
for k in K:
    single_space = dict()
    for s in S:
        single_space[s] = call_price(s, k, ttm, r, sigma)
    call_space[k] = pd.Series(single_space)
call_space
```

```python
# 데이터프레임화
df_call_space = pd.DataFrame(call_space)
df_call_space
```

#### 풋옵션가격 곡면

```python
def put_price(S, K, ttm, r, sigma):
    d1 = (np.log(S/K) + (r + sigma**2*0.5) * ttm) / (sigma * np.sqrt(ttm))
    d2 = (np.log(S/K) + (r + sigma**2*0.5) * ttm) / (sigma * np.sqrt(ttm))
    val = K * np.exp(-r * ttm) * stats.norm.cdf(-d2,0.0,1.0) - (S * stats.norm.cdf(-d1,0.0,1.0))
    return val
```

```python
K = np.arange(250,350,5)
S = np.arange(200,400,10)
K,S = np.meshgrid(K,S)
Z = put_price(S,K,ttm,r,sigma)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(S,K,Z,cmap=cm.winter,
                       linewidth=1,antialiased=True,alpha=0.8)

ax.set_xlabel('기초자산(Underlying)')
ax.set_ylabel('행사가격(Strike)')
ax.set_zlabel('풋옵션가격(Put Price)')
ax.set_title('풋옵션 가격 곡면')

fig.colorbar(surf, shrink=0.5, aspect=5)
```

### 옵션 그릭스 그래프 그리기

#### 콜옵션 델타 그래프

```python
def call_delta(S,K,ttm,r,sigma):
    d1 = (np.log(S/K) + (r+sigma**2*0.5)*ttm) / (sigma*np.sqrt(ttm))
    val = stats.norm.cdf(d1, 0.0, 1.0)
    return val
```

```python
K = np.arange(250,350,5)
S = np.arange(200,400,10)
K,S = np.meshgrid(K,S)
Z = call_delta(S,K,0.38,0.02,0.30)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(S,K,Z,cmap=cm.coolwarm,
                       linewidth=1,antialiased=True,alpha=0.8)

ax.set_xlabel('기초자산(Underlying)')
ax.set_ylabel('행사가격(Strike)')
ax.set_zlabel('델타(Delta)')
ax.set_title('델타 곡면')

fig.colorbar(surf, shrink=0.5, aspect=5)
```

#### 감마 그래프

```python
def ndx(x):
    return (np.exp(-1*x**2*0.5)) / np.sqrt(2*np.pi)
```

```python
# 감마는 콜과 풋의 공식이 동일하므로 구분자 없이 정의
def gamma(S,K,ttm,r,sigma):
    d1 = (np.log(S/K) + (r+sigma**2*0.5)*ttm) / (sigma*np.sqrt(ttm))
    val = (ndx(d1)) / (S*sigma*np.sqrt(ttm))
    return val
```

```python
K = np.arange(300,305)
S = np.arange(200,400,10)
K,S = np.meshgrid(K,S)
Z = call_delta(S,K,0.38,0.02,0.30)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(S,K,Z,cmap=cm.summer,
                       linewidth=1,antialiased=True,alpha=0.8)

ax.set_xlabel('기초자산(Underlying)')
ax.set_ylabel('행사가격(Strike)')
ax.set_zlabel('감마(Gamma)')
ax.set_title('감마 곡면')

fig.colorbar(surf, shrink=0.5, aspect=5)
```

#### 콜옵션 세타 그래프

```python
def call_theta(S,K,ttm,r,sigma):
    d1 = (np.log(S/K) + (r+sigma**2*0.5)*ttm) / (sigma*np.sqrt(ttm))
    d2 = (np.log(S/K) + (r-sigma**2*0.5)*ttm) / (sigma*np.sqrt(ttm))
    val = -1*((S*ndx(d1)*sigma) / (2*np.sqrt(ttm))) - r*K*np.exp(-r*ttm)*stats.norm.cdf(d2,0.0,1.0)
    return val
```

+ 세타만의 특이한 점
  - 잔존만기가 점점 0에 가까워지면서 그래프가 변하는 모습을 보여주기 위해 X축 값이 양수에서 0으로 거꾸로 흘러가게 만들어야 함

```python
T = np.arange(1.0,0.0,-0.01)
S = np.arange(350,355)
T,S = np.meshgrid(T,S)
Z = call_theta(S,350,T,0.02,0.30)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(T,S,Z,cmap=cm.summer,
                       linewidth=1,antialiased=True,alpha=0.8)

ax.set_xlabel('잔존만기(Time to Maturity)')
ax.set_ylabel('기초자산(Underlying)')
ax.set_zlabel('세타(Theta)')
ax.invert_xaxis() # X축을 1부터 0의 순서로 거꾸로 출력
ax.set_title('세타 곡면')

fig.colorbar(surf, shrink=0.5, aspect=5)
```

#### 베가 그래프

```python
# 베가는 콜과 풋의 구분이 없음
def vega(S,K,ttm,r,sigma):
    d1 = (np.log(S/K) + (r+sigma**2*0.5) * ttm) / (sigma*np.sqrt(ttm))
    d2 = (np.log(S/K) + (r-sigma**2*0.5) * ttm) / (sigma*np.sqrt(ttm))
    val = (S*np.sqrt(ttm)*ndx(d1))
    return val
```

+ 베가는 변동성의 변화에 따른 옵션가격의 변화를 나타냄

```python
V = np.arange(0.05,0.95,0.01)
S = np.arange(350,352)
V,S = np.meshgrid(V,S)
Z = vega(S, 350, 0.38, 0.02, V)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(V,S,Z,cmap=cm.autumn,
                       linewidth=1,antialiased=True,alpha=0.8)

ax.set_xlabel('변동성(Volatility)')
ax.set_ylabel('기초자산(Underlying)')
ax.set_zlabel('베가(Vega)')
ax.set_title('베가 곡면')

fig.colorbar(surf, shrink=0.5, aspect=5)
```

+ 변동성이 아주 낮은(10%미만) 구간에서 베가가 급격히 올라가고, 변동성이 증가함에 따라 베가가 점점 감소함
