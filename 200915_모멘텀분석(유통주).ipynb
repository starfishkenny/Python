```python
import sqlite3
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 데이터베이스에서 필요한 데이터를 불러온다
dbname='Naver_stock.db'
def selData(whr=' where SIDX=0'):
    with sqlite3.connect(dbname) as conn:
        cur=conn.cursor()
        sql='select SDAY,OPEN_P,HIGH_P,LOW_P,CLOSE_P,VOL from 현대백화점 '+whr
        cur.execute(sql)
        rows=cur.fetchall()
        cur.close()
    return rows
```

```python
# sqlite3에서 실행할 쿼리문 작성
whr='where SDAY>20200101 order by SDAY desc limit 250'
res=selData(whr)
res
```

```python
# 데이터프레임화
cols = ['SDAY','OPEN_P','HIGH_P','LOW_P','CLOSE_P','VOL']
df = pd.DataFrame(data=res,columns=cols)
df.head(3)
```

```python
# 종가와 시가 비교
sels=['CLOSE_P','OPEN_P']
df[sels].plot()
plt.xlim(25,100)
plt.ylim(50000,80000)
plt.show()
```
